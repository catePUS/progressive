```python
import pandas as pd

rat= pd.read_csv('ratings.csv')

a = rat.groupby(['userId']).count().reset_index()
a = a.loc[a.rating>100]

b = rat.loc[rat['userId'].isin(a.userId)]
b = b.groupby('userId').agg({'timestamp': ['max', 'min']})
b['mean_timestamp']= b['timestamp']['max'] - b['timestamp']['min']
mean_timestamp = b['mean_timestamp'].mean()
print(mean_timestamp)
b.head(1000)
```
